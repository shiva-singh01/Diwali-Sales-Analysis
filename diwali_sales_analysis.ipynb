```python
# most amount spend by gender and total sum 

sales_gn = df.groupby(['Sex'], as_index = False)['Amount'].sum().sort_values(by= 'Amount', ascending = False)
```

```python
#most amount spend by gender and total sum descending

sales_gn = df.groupby(['Sex'], as_index= False)['Amount'].sum().sort_values(by= 'Amount', ascending = True)
```

```python
sns.barplot(x = 'Sex', y = 'Amount', data= sales_gn)
```
